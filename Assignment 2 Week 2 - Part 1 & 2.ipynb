### Bar Chart

1. Display a bar chart of the popularity of Smart Phones

   Sample Data:

   Smart Phones :Apple,Samsung,Oppo,Redmi,Vivo

   Popularity:23.3, 52.6, 11.2,8, 6.7
   




2. Display a horizontal bar chart for the above question.





3. Display a bar chart for the above question using different color for each bar and attach a text label above each bar displaying its popularity.

```python
import matplotlib.pyplot as plt

smartphones = ['Apple', 'Samsung', 'Oppo', 'Redmi', 'Vivo']
popularity = [23.3, 52.6, 11.2, 8, 6.7]
```

```python
plt.bar(smartphones, popularity)
plt.title('Popularity of Smart Phones')
plt.xlabel('Smart Phones')
plt.ylabel('Popularity (%)')
plt.show()
```

```python
plt.barh(smartphones, popularity)
plt.title('Popularity of Smart Phones')
plt.ylabel('Smart Phones')
plt.xlabel('Popularity (%)')
plt.show()
```

```python
plt.bar(smartphones, popularity, color=['blue', 'green', 'red', 'orange', 'purple'])
for i in range(len(smartphones)):
    plt.text(i, popularity[i], str(popularity[i]), ha='center', va='bottom')
plt.show()
```

### Read all product sales data and show it using a multiline plot.Display the number of units soldper month for each product using multiline plots. (i.e., Separate Plotline for each product).

```python
import pandas as pd 
data = pd.read_csv('data_document.csv')
data
```

```python
plt.figure(figsize=(10, 6)) 
plt.plot(data.month_number, data.facecream, label='Face Cream')
plt.plot(data.month_number, data.facewash, label='Face Wash')
plt.plot(data.month_number, data.toothpaste, label='Toothpaste')
plt.plot(data.month_number, data.bathingsoap, label='Bathing Soap')
plt.plot(data.month_number, data.shampoo, label='Shampoo')
plt.plot(data.month_number, data.mosturizer, label='Moisturizer')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.title('Number of Units Sold per Month for Each Product')
plt.legend() 
plt.show()  
```

### Perform various data cleaning and data visualization operations on the previously given dataset. (Present visualization in at least five different plots.)

```python
df=data
df
```

```python
df.describe()
```

```python
df.shape
```

```python
df.info()
```

```python
#checking missing values
missing_values = df.isnull().sum()
print("Missing Values:\n", missing_values)
```

```python
df = df.dropna()
```

Bar Plot

```python
product_totals = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'mosturizer']].sum()
plt.figure(figsize=(10, 6))
product_totals.plot(kind='bar', color='blue')
plt.xlabel('Product')
plt.ylabel('Total Units')
plt.title('Total Units Sold by Product')
plt.show()
```

Line Plot

```python
months = df['month_number']
monthly_totals = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'mosturizer']].sum(axis=1)
plt.figure(figsize=(10, 6))
plt.plot(months, monthly_totals, marker='*', linestyle='-', color='orange')
plt.xlabel('Month')
plt.ylabel('Total Units')
plt.title('Monthly Sales Trend')
plt.show()
```

Pie Chart

```python
product_profits = df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'mosturizer']].sum()
plt.figure(figsize=(10, 6))
plt.pie(product_profits, labels=product_profits.index,autopct='%1.1f%%',startangle=90)
plt.title('Total Profit by Product')
plt.axis('equal')
plt.legend()
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['facecream'], df['facewash'], color='red', label='Face Wash')
plt.scatter(df['facecream'], df['toothpaste'], color='green', label='Toothpaste')
plt.scatter(df['facecream'], df['bathingsoap'], color='blue', label='Bathing Soap')
plt.scatter(df['facecream'], df['shampoo'], color='purple', label='Shampoo')
plt.scatter(df['facecream'], df['mosturizer'], color='orange', label='Mosturizer')
plt.xlabel('Face Cream Units Sold')
plt.ylabel('Units Sold')
plt.title('Product-wise Sales Comparison')
plt.legend()
plt.show()
```

Box Plot

```python
plt.figure(figsize=(10, 6))
df[['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'mosturizer']].boxplot()
plt.xlabel('Product')
plt.ylabel('Units Sold')
plt.title('Product Sales Distribution')
plt.show()
```

### Plot Mean and Standard Deviation in Pandas

```python
mean_values = df.mean()
std_values = df.std()
```

```python
stastics = pd.DataFrame({'Mean': mean_values, 'Standard Deviation': std_values})
```

```python
stastics.plot(kind='bar', figsize=(10, 6))
plt.xlabel('Product')
plt.ylabel('Value')
plt.title('Mean and Standard Deviation for Each Product')
plt.show()
```
